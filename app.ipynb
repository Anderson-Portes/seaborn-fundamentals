```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
exercise = sns.load_dataset('exercise')
flights = sns.load_dataset('flights')
diamonds = sns.load_dataset('diamonds').sample(10000)
```

```python
tips
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips, hue='time',size='size', palette='coolwarm')
```

```python
sns.histplot(tips['tip'], bins=15, kde=True)
```

```python
sns.displot(tips['tip'], bins=15, kde=True)
```

```python
sns.barplot(x='sex',y='tip', data=tips, palette='YlGnBu')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='sex', palette='YlGnBu')
```

```python
sns.stripplot(x='day',y='tip',data=tips, hue='sex', dodge=True, palette='coolwarm')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',fill=True)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='hex')
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```

```python
titanic.select_dtypes(['number']).corr()
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(), annot=True,cmap='coolwarm')
```
